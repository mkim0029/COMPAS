```python
import os, sys    # for handling paths
import h5py as h5  #for handling data format
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
def restrict(df, col, lower_limit, upper_limit):
    mask = (df[col] >= lower_limit) & (df[col] <= upper_limit)
    return df[mask]
```

```python
directories = ['1Msun+1Msun', '10Msun+2Msun', '3Msun+3Msun', '6Msun+2Msun', '7Msun+1Msun']
for d in directories:
    #df1 = pd.read_csv('BSE_Detailed_Output_0.csv', sep = ',', header = 2)
    df2 = pd.read_csv(d + '/SSE_Detailed_Output_0.csv', sep = ',', header = 2)
    
    #df1 = restrict(df1, 'Stellar_Type(1)', 0, 1) #MS stars only
    #df1 = restrict(df1, 'Stellar_Type(2)', 15, 15) #merger product only
    df2 = restrict(df2, 'Stellar_Type', 0, 1) #MS stars only

    plt.scatter(df2['          Age           '], df2['          Mdot          '], s = 0.5, label = str(round(df2['          Mass          '][0], 1)))
    
plt.xlabel('Age (Myr)')
plt.ylabel('Mdot (Msun/yr)')
plt.title('Mass loss rate')
plt.xlim((0,200))
plt.ticklabel_format(useOffset=False)
plt.legend()
```
